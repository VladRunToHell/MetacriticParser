Отсюда убираем всё, что имеет меньше 10 записей (3DO, GG, PCFX, SCD, TG16, WS). Остаток - 16697. Остаток на выброс - 1652

Также убираем записи, которые не имеют названия (Name = NaN). Остаток - 16695. Остаток на выброс - 1650

```python
data = data.drop(data[data['Name'].isna()].index)
```

Далее проверим руками, есть ли на Metacritic оценки для платформ, которые не имеют оценки ни на одну игру

Для следующих платформ оценок нет: Atari 2600 (2600), GameBoy (GB), Sega Mega Drive (GEN), NES, Neo-Geo AES (NG), Sega Saturn (SAT), Super NES (SNES). Их придётся выкинуть: остаток - 15915, остаток на выброс - 870

Из всего списка, оценки возможно найти только для Nintendo 64.

```python
platforms = ["2600", "GB", "GEN", "NES", "NG", "SAT", "SNES"]
for platform in platforms:
    data = data.drop(data[data['Platform']==platform].index)
```

```python
data.groupby(by="Platform").count()
```
